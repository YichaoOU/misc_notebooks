```python
import pandas as pd
import glob
```

```python
df_list = []
for i in range(1,6):
    folder = "HiC_Hist1/HiC_top"+str(i) + "/"
    enriched_motifs = folder+"HiC_top"+str(i) + ".enriched_motifs.csv"
    m = pd.read_csv(enriched_motifs)
    m = m[m['State'] == "E7"]
    if m.shape[0] == 0:
        print (enriched_motifs,"no promoter motifs")
        continue
    m = m.sort_values('P-value').head(n=2)
    m['Hub'] = "Hub" + str(i)
    df_list.append(m)
m_df = pd.concat(df_list)
m_df['Motif'] = list(map(lambda x:x.split("_")[0].replace('mm9.',''),m_df['Motif'].tolist()))
m_df.to_csv("HiC_hubs.promoter.top2.motifs.csv",index=False)
```

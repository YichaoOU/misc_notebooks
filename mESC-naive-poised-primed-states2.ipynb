```python
my_genes = {}
count = 0
for k in gset1.keys():
    for g in gset1[k].index.tolist():
        if not isinstance(g, str):
            continue
        if not g in my_genes:
            my_genes[g] = []
        my_genes[g].append(real_name[count])
    count+=1
```

```python
names = gam[cols[0]].tolist()+gam[cols[1]].tolist()
names = list(set(names))
```

```python
genes_final = {}
for g in names:
    if g == 0:
        continue
    temp = g.split(",")
    for m in my_genes:
        for t in temp:
            t = t.replace(" ","")
            if m.lower() == t.lower():
                genes_final[t] = my_genes[m]
```

```python
for i in df.index:
    for j in df.columns:
        value = df.at[i,j]
        if i != j:
            if value <= 0.5:
                endi = int(i.split("-")[-1])
                endj = int(j.split("-")[-1])
                
                dist = abs(endj-endi)
                if dist <= 90000:
                    df.set_value(i,j,1)
                else:
                    dist = dist / 30000
                    df.set_value(i,j,dist)
            else:
                df.set_value(i,j,1-value)
```

```python
import numpy as np
from sklearn.cluster import KMeans,AgglomerativeClustering
cluster_hc = AgglomerativeClustering(n_clusters=100,affinity="precomputed",linkage="average")
cluster_hc.fit(df)
predicted=cluster_hc.labels_
print np.unique(predicted)
current_palette = sns.color_palette("hls", len(np.unique(predicted)))
a = dict(zip(np.unique(predicted), current_palette))
color_map = map(lambda x:a[x],predicted)
color3 = pd.DataFrame()
color3['name'] = df.index.tolist()
color3['HC-clusters'] = color_map
color3 = color3.set_index("name")
myColors = sns.cubehelix_palette(20)[3:]
cmap = LinearSegmentedColormap.from_list('Custom', myColors, len(myColors))
df_c_2 = pd.concat([color2,color3],axis=1)
df_c_2.head()
sns.clustermap(data[data>0.5].iloc[700:830,700:830],row_cluster=False,col_cluster=False,row_colors = df_c_2,col_colors = df_c_2,cmap = cmap)
```

```python
import networkx as nx
```

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
g = nx.read_gml("Hist1_D_norm_result.network.gml")
```

```python
for n in g.nodes:
    print (n.degree)
```

```python
my_dict = g.degree
```

```python
for x in my_dict:
    print (x[0])
```

```python
df = [[x[0],x[1]] for x in g.degree]
```

```python
df = pd.DataFrame(df)
```

```python
df['start'] = df[0].apply(lambda x:x.split("-")[0].split(":")[-1]).astype(int).tolist()
```

```python
temp = df[(df['start'] < 24100000) & (df['start'] > 21700000)]
```

```python
temp.head()
```

```python
temp.columns = ['label','degree','start']
```

```python
sns.lmplot(x="start",y='degree',data=temp.sort_values('start'))
```

```python
sns.kdeplot(data=temp.sort_values('start')['degree'])
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.kdeplot(data=temp.sort_values('start')['degree'])
sns.lmplot(x="start",y='degree',data=temp.sort_values('start'),regfit=False)
```

```python
sns.lmplot(x="start",y='degree',data=temp.sort_values('start'),fit_reg=False)
```

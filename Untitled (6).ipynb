```python
import networkx as nx
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("SPRITE_HIST1_degree_df_annotation.csv",index_col=0)
```

```python
temp = df[(df['start'] <= 24100000) & (df['end'] >= 21700000)]
```

```python
temp['Number of contacts'].describe().to_csv("SPRITE-hist1-degree-stats.csv")
```

```python
def degree_distplot(df,outfile_name):
	f,ax1 = plt.subplots()
	sns.distplot(df['Number of contacts'],kde=False,ax=ax1)
	ax1.set_xlabel('Number of contacts')
	plt.title("Number of contacts histogram")
	plt.savefig(outfile_name)
	return [outfile_name]
```

```python
degree_distplot(temp,"SPRITE-hist1-degree-dist.png")
```

```python
lm = sns.lmplot(x="start",y='Number of contacts',data=temp.sort_values('start'),fit_reg=False)
axes = lm.axes
axes[0,0].set_ylim(5640,5680)
```

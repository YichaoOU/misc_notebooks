```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import gc
from stacking_clf import mixer_reg,xgb_clf

df_train = pd.read_csv("training_all_features.csv")
x_train = df_train.drop(['parcelid', 'logerror', 'transactiondate'], axis=1)
y_train = df_train['logerror']
```

```python
import xgboost as xgb
dtrain = xgb.DMatrix( x_train, label=y_train)
```

```python
dtrain.feature_names
```

```python
mapper = {'f{0}'.format(i): v for i, v in enumerate(dtrain.feature_names)}
```

```python
xgb_params = {"objective": "reg:linear"}
```

```python
gbdt = xgb.train(xgb_params, dtrain, 1000)
```

```python
from sklearn.svm import SVR
from sklearn.linear_model import Lasso,ElasticNet
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor,ExtraTreesRegressor
from sklearn.linear_model import Ridge,LinearRegression,HuberRegressor,PassiveAggressiveRegressor,BayesianRidge,ARDRegression,Lars,LassoLars,LassoLarsIC,MultiTaskLasso,OrthogonalMatchingPursuit,RANSACRegressor,TheilSenRegressor
```

```python
from sklearn.linear_model import Ridge, RidgeCV, ElasticNet, ElasticNetCV, Lasso, LassoCV, LassoLars, LassoLarsCV, RandomizedLasso
```

```python
model_rlasso = RandomizedLasso(alpha='bic', verbose=False, n_resampling=500, random_state=1001, n_jobs=1)
```

```python
model_rlasso.fit(x_train, y_train)
```

```python
coef = pd.DataFrame(model_rlasso.scores_, columns = ['RandomizedLasso_score'])
coef['Feature'] = x_train.columns
coef['Relative score'] = coef['RandomizedLasso_score'] / coef['RandomizedLasso_score'].sum()
coef = coef.sort_values('Relative score', ascending=False)
coef = coef[['Feature', 'RandomizedLasso_score', 'Relative score']]
coef.to_csv("feature_importance_randomizedlasso.csv", index=False)
```

```python
model_rlasso = Lasso(alpha=0.01, warm_start=True,max_iter=1000,normalize=True,fit_intercept=True)
model_rlasso.fit(x_train, y_train)
coef = pd.DataFrame(model_rlasso.coef_ , columns = ['Regular_lasso'])
coef['Feature'] = x_train.columns
coef['Relative score'] = coef['Regular_lasso'] / coef['Regular_lasso'].sum()
coef = coef.sort_values('Relative score', ascending=False)
coef = coef[['Feature', 'Regular_lasso', 'Relative score']]
coef.to_csv("feature_importance_regularLasso.csv", index=False)
```

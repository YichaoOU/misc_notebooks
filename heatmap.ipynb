```python
import pandas as pd

data = pd.read_csv("gene_TPM_matrix.csv",index_col=0)

my_list = map(lambda x:x.strip(),open("union_of_genes.list").readlines())
```

```python
df = data[data['gene_name'].isin(my_list)]
```

```python
import seaborn as sns
```

```python
df[df.columns[1:]]
```

```python
data[data['gene_name'].isin(my_list)]
```

```python
%matplotlib inline
sns.set(color_codes=True)
sns.set(font_scale=0.5)
data2 = df[df.columns[1:]]
data2.index = data[data['gene_name'].isin(my_list)]['gene_name']
plot = seaborn_map = sns.clustermap(data2, row_cluster=True, col_cluster=True,standard_scale =0,col_colors=["red","red","red","green","green","green"],figsize=(10,15))
plt.setp(plot.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plot.savefig("heatmap.png",dpi=1000)

```

```python
import matplotlib.pyplot as plt
plt.setp(plot.ax_heatmap.get_yticklabels(), rotation=0)
sns.plt.show()
```

```python
%matplotlib inline
sns.set(color_codes=True)
plot = seaborn_map = sns.clustermap(df[df.columns[1:]], cmap=sns.diverging_palette(150, 0, s=80, l=55, n=9,as_cmap=True),row_cluster=True, col_cluster=True,standard_scale =0,col_colors=["red","red","red","green","green","green"])
plot.savefig("heatmap.png")

```

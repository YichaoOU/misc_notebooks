```python
import networkx as nx
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("HiC_Hist1/HiC_C2/HiC_HIST1_degree_df_annotation.csv",index_col=0)
```

```python
import matplotlib.pyplot as plt
temp = df[(df['start'] <= 24100000) & (df['end'] >= 21700000)]
temp['Number of contacts'].describe().to_csv("HiC-hist1-degree-stats.csv")
def degree_distplot(df,outfile_name):
	f,ax1 = plt.subplots()
	sns.distplot(df['Number of contacts'],kde=False,ax=ax1)
	ax1.set_xlabel('Number of contacts')
	plt.title("Number of contacts histogram")
	plt.savefig(outfile_name)
	return [outfile_name]
degree_distplot(temp,"HiC-hist1-degree-dist.png")
```

```python
df = pd.read_csv("SLICE_Hist1/SLICE_C2/SLICE_HIST1_degree_df_annotation.csv",index_col=0)
temp = df[(df['start'] <= 24100000) & (df['end'] >= 21700000)]
temp['Number of contacts'].describe().to_csv("SLICE-hist1-degree-stats.csv")
degree_distplot(temp,"SLICE-hist1-degree-dist.png")
```

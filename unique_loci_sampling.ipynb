```python
import pandas as pd
import random
df = pd.read_csv("Dnorm.hist1.MP.test.csv",index_col=0)
```

```python
df['W1'] = list(map(lambda x:x.split("%%")[0],df.index.tolist()))
df['W2'] = list(map(lambda x:x.split("%%")[1],df.index.tolist()))
unique_windows = df['W1'].tolist()+df['W2'].tolist()
```

```python
def get_motifs(x):
    return x.split("%%")
```

```python
[m1,m2] = get_motifs(df.index[0])
```

```python
index_list = df.index
```

```python
df.loc[index_list][(df['W1']!=m1) & (df['W2']!=m2)]
```

```python
def iterative_sampling(df):
    index_list = df.index.tolist()
    random_pick=random.choice(index_list)
    [m1,m2] = get_motifs(random_pick)
    selected = [random_pick]
    while True:
        index_list = df.loc[index_list][(df['W1']!=m1) & (df['W2']!=m2)].index.tolist()
        if len(index_list) <= 1:
            break
        random_pick=random.choice(index_list)
        [m1,m2] = get_motifs(random_pick)
        selected.append(random_pick)
    return selected 
    
```

```python
selected = iterative_sampling(df)
```

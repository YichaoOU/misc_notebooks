```python

import pandas as pd


df = pd.read_csv("motif_occurrence.csv")


motif_pairs = map(lambda x:x.replace("_1","").replace("_2",""),df.motif_name.tolist())

motif_pairs = list(set(motif_pairs))
print len(motif_pairs)

def parse_occ(file):
	my_dict = {}
	flag = True
	with open(file) as f:
		for line in f:
			if flag:
				flag = False
				continue
			line = line.strip().split(",")
			seq_name = line[0]
			motif_name = line[1]
			if not my_dict.has_key(seq_name):
				my_dict[seq_name] = {}
			if not my_dict[seq_name].has_key(motif_name):
				my_dict[seq_name][motif_name] = 1
	return my_dict

# df = df.set_index('seq_name')
res = pd.DataFrame(columns = motif_pairs)
pair = "EE.pos.cHMM_pairs.tsv"
df = parse_occ("motif_occurrence.csv")
with open(pair) as f:
	for line in f:
		line = line.strip().split()
		my_res = []
		for mp in motif_pairs:
			flag1 = False
			flag2 = False
			try:
				# f1 = df.at[line[0],'motif_name']
				r1m1 = df[line[0]][mp+"_1"]
				print "r1m1",r1m1
				r2m2 = df[line[1]][mp+"_2"]
				print "r2m2",r2m2
				flag1 = True
			except:
				flag1 = False
			try:
				r1m2 = df[line[0]][mp+"_2"]
				print "r1m2",r1m2
				r2m1 = df[line[1]][mp+"_1"]
				print "r2m1",r2m1
				flag2 = True
			except:
				flag2 = False
			if flag1 or flag2:
				my_res.append(1)
			else:
				my_res.append(0)
		res.loc[",".join(line)] = my_res







```

```python
for i in df.index:
    for j in df.columns:
        value = df.at[i,j]
        if i != j:
            if value <= 0.5:
                endi = int(i.split("-")[-1])
                endj = int(j.split("-")[-1])
                
                dist = abs(endj-endi)
                if dist <= 90000:
                    df.set_value(i,j,1)
                else:
                    dist = dist / 10000
                    df.set_value(i,j,dist)
            else:
                df.set_value(i,j,1-value)
```

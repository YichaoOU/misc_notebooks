```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("SLICE_05_mmc2_all_overlaped_interactions.csv",index_col=0)
```

```python
tdf = df.transpose()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
y1 = PCA(2).fit_transform(tdf)
data1 = pd.DataFrame(y1,columns=["pc1",'pc2'])
sns.lmplot(x="pc1", y="pc2", data=data1,fit_reg=False,scatter_kws={'alpha':0.3})
```

```python
y1 = TSNE().fit_transform(tdf)
data1 = pd.DataFrame(y1,columns=["p1",'p2'])
sns.lmplot(x="p1", y="p2", data=data1,fit_reg=False,scatter_kws={'alpha':0.3})
```

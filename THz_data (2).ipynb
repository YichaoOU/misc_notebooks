```python
def find_sample_name(text):
	text = text.split("\n")
	temp = []
	for t in text:
		if t == "":
			continue
		t = t.split()
		temp1 = []
		for i in t:
			if i == "":
				continue
			temp1.append(i)
		temp.append("_".join(temp1).replace(",",""))
	return "_".join(temp[::-1])


```

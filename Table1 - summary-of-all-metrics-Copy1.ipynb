```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
enrichment = pd.read_csv("enrichment_MS_summary.csv",usecols = [0,1,2,3,6],header = None)
enrichment.columns = ["Groups","Sensitivity","Specificity","Accuracy","Num_motifs"]
enrichment.set_index("Groups",inplace=True)
enrichment['Specificity'] = 1- enrichment['Specificity']
enrichment['Accuracy'] = 1- enrichment['Accuracy']

greedy = pd.read_csv("greedy_result.csv",usecols = [0,1,2,3,6],header = None)
greedy.columns = ["Groups","Sensitivity","Specificity","Accuracy","Num_motifs"]
greedy.set_index("Groups",inplace=True)
greedy['Specificity'] = 1- greedy['Specificity']
greedy['Accuracy'] = 1- greedy['Accuracy']

RILP = pd.read_csv("RILP_result.csv",usecols = [0,1,2,3,6],header = None)
RILP.columns = ["Groups","Sensitivity","Specificity","Accuracy","Num_motifs"]
RILP.set_index("Groups",inplace=True)
RILP['Specificity'] = 1- RILP['Specificity']
RILP['Accuracy'] = 1- RILP['Accuracy']
Tabu_PNPSC = pd.read_csv("Tabu_PNPSC_result.csv",usecols = [0,1,2,3,6],header = None)
Tabu_PNPSC.columns = ["Groups","Sensitivity","Specificity","Accuracy","Num_motifs"]
Tabu_PNPSC.set_index("Groups",inplace=True)
Tabu_PNPSC['Specificity'] = 1- Tabu_PNPSC['Specificity']
Tabu_PNPSC['Accuracy'] = 1- Tabu_PNPSC['Accuracy']
random_selection = pd.read_csv("Random_motifs_result.csv",usecols = [0,1,2,3,6],header = None)
random_selection.columns = ["Groups","Sensitivity","Specificity","Accuracy","Num_motifs"]
random_selection.set_index("Groups",inplace=True)
random_selection['Specificity'] = 1- random_selection['Specificity']
random_selection['Accuracy'] = 1- random_selection['Accuracy']

```

```python
table_ForeCov = pd.concat([enrichment["Sensitivity"],greedy["Sensitivity"],RILP["Sensitivity"],Tabu_PNPSC["Sensitivity"]],axis=1)
table_ForeCov.columns = ['Enrch','Grdy','RILP','TABU']
table_ForeCov = table_ForeCov*100
table_ForeCov = table_ForeCov.round(1)
table_ForeCov['max'] = table_ForeCov.max(axis=1)
table_ForeCov = table_ForeCov.fillna("NA")
table_ForeCov = table_ForeCov.sort_values('Enrch')
```

```python
header = """

\\begin{table}[]
\centering
\def\\arraystretch{1.3}
\caption{Foreground coverage(\%) comparison across 55 factors}
\label{my-label}
\\begin{tabular}{|l|c|c|c|c|}
\hline
Factor  & Enrch               & Grdy                & RILP                & TABU                \\\\ \hline


"""

def print_table(file):
	lines = open(file).readlines()
	num_max_count = [0,0,0,0]
	for line in lines[1:]:
		line = line.strip().split("\t")
		max_value = line[-1]
		outline = line[0] + "\t\t& "
		count = 0
		for ele in line[1:-1]:
			if ele == "NA":
				outline += ele + "\t\t& "
				continue
			if float(ele) == float(max_value):
				outline += "{\\textbf{my_value}}".replace("my_value",max_value) + "\t\t& "
# 				print count,line
				num_max_count[count] += 1
			else:
				outline += ele + "\t\t& "
			count +=1
		outline = outline[:-2]
		outline += "\\\\ \hline"
		print outline
	outline = map(lambda x:str(x) + "\t\t& ",num_max_count)
# 	print outline
	outline = "Winning Times 		& "+ "".join(outline)[:-2] + "\\\\ \hline"
	print outline
print_table("table_forecov.tsv")
```

```python
table_ForeCov = pd.concat([enrichment["Specificity"],greedy["Specificity"],RILP["Specificity"],Tabu_PNPSC["Specificity"]],axis=1)
table_ForeCov.columns = ['Enrch','Grdy','RILP','TABU']
table_ForeCov = table_ForeCov*100
table_ForeCov = table_ForeCov.round(1)
table_ForeCov['min'] = table_ForeCov.min(axis=1)
table_ForeCov = table_ForeCov.fillna("NA")
table_ForeCov = table_ForeCov.sort_values('Enrch',ascending=False)
table_ForeCov.to_csv("temp.tsv",sep="\t")
# print header
print_table("temp.tsv")
```

```python
import pandas as pd
df = pd.read_csv("GSE64881_segmentation_at_30000bp.passqc.multibam.txt",sep="\t")
def my_index(df):
	return df['chrom'] + ":" + df['start'].astype(str) + "-" + df['stop'].astype(str)


df.index = my_index(df)
df = df.drop(['chrom','start','stop'],axis=1)
df.head()
```

```python
df.shape
```

```python
df = df.transpose()
df.head()
```

```python
slice05_pairs = pd.read_csv("chromatin_interaction_pairs_of_features.csv")
slice05_pairs.head()
```

```python
slice05_pairs.shape
```

```python
feature_list = slice05_pairs.name + "," + slice05_pairs.variable
```

```python
import numpy as np
np.zeros((2,3))
```

```python
my_df = pd.DataFrame(np.zeros((df.shape[0],len(feature_list))))
my_df.columns = feature_list
my_df.index = df.index
my_df.head()
```

```python
for j in feature_list:
    l1,l2 = j.split(",")
    for i in my_df.index:
        l1_value = df.at[i,l1]
        l2_value = df.at[i,l2]
        if l1_value and l2_value:
            my_df.set_value(i,j,1)
```

```python
my_df.to_csv("GAM_05_single_cell_pairs.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
y1 = PCA(2).fit_transform(my_df)
data1 = pd.DataFrame(y1,columns=["pc1",'pc2'])
sns.lmplot(x="pc1", y="pc2", data=data1,fit_reg=False,scatter_kws={'alpha':0.3})
```

```python
slide = frame[200:300,:,:]
```

```python
import numpy as np
from scipy import stats
def detect_yellow(frame):
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower = np.array([23,60,60], dtype=np.uint8)
    upper = np.array([32, 255, 255], dtype=np.uint8)
    mask = cv2.inRange(hsv, lower, upper)
    x,y = mask.nonzero()
#     slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    print (stats.linregress(x,y))
    rate = np.count_nonzero(mask) / (mask.shape[0] * mask.shape[1])
    print (rate)
    return rate,mask  
r,m = detect_yellow(slide)
plt.imshow(m,cmap='gray')
```

```python
def detect_white(frame):
    sensitivity = 50
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower = np.array([0,0,255-sensitivity], dtype=np.uint8)
    upper = np.array([180, sensitivity, 255], dtype=np.uint8)
    mask = cv2.inRange(hsv, lower, upper)
    x,y = mask.nonzero()
    print (x)
    print (y)
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    print (slope, intercept, r_value, p_value, std_err)
    print (slope*50+intercept)
#     print (stats.linregress(y,x))
    rate = np.count_nonzero(mask) / (mask.shape[0] * mask.shape[1])
    return rate,mask  
r,m = detect_white(slide)
plt.imshow(m,cmap='gray')
```

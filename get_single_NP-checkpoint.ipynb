```python
temp = gam2.columns.tolist()
temp[0] = 'locus2'
gam2.columns = temp
gam2.head()
```

```python
temp = gam2.columns.tolist()
new_names = []
for i in temp:
    if "R1" in i:
        new_names.append(i.replace("R1","R2"))
    elif "R2" in i:
        new_names.append(i.replace("R2","R1"))
    else:
        new_names.append(i)
gam2.columns = new_names
gam2.head()        
```

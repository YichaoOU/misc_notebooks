```python
import pandas as pd
mapping={}
mapping["DOGSIFH100529.2"] = "PSO3"
mapping["DOGSIFH100529.4"] = "LMP"
mapping["DOGSIFH100529.6"] = "PSO3"
mapping["DOGSIFH100529.8"] = "LMP"
mapping["PLATYPUS100611.5"] = "PSO2"
mapping["PLATYPUS100611.7"] = "PSO3"
mapping["DOGSIFH100529.3"] = "LMP"
mapping["DOGSIFH100529.5"] = "PSO3"
mapping["DOGSIFH100529.7"] = "LMP"
mapping["LEOPARD100805.7"] = "PSO2"
mapping["PLATYPUS100611.6"] = "PSO2"
```

```python
my_dict = {}
my_dict['LMP'] = ["STAG3L5P-PVRIG2P-PILRB","SFT2D2","SMG1","KPNA2","STAG3","BPTF"]
my_dict['PSO2'] = ["HHLA1","EMC3","SPINT2","STAG3L5P-PVRIG2P-PILRB","C19orf33","CIDEC","SEPT14","CRISPLD2","FBXO25","PRR11","SOX13","KPNA2","STAG3","BPTF","BCL7C","HHLA1"]
my_dict['PSO3'] = ["TMEM165","TBC1D17","DNTTIP2","RASEF","ZNF207","KCNK6","MRPS22","STK33","STAG3","ZNF480","LCMT2","FCRL5","TMEM165"]
```

```python
def get_all_data(file_list):
    d = []
    for f in file_list:
        df = pd.read_csv(f,sep="\t")
        columns = ['GeneName1','GeneName2','GeneExpr1','GeneExpr2','GeneExpr_Fused','EricScore']
        df = df[columns]
        df['File'] = f.replace(".results.filtered.tsv",'')
        d.append(df)
    return d
PSO3_df_list = []
for k in mapping:
    file = k+".results.filtered.tsv"
    value = mapping[k]
    if value != "PSO3":
        continue
    print (file,value)
    PSO3_df_list.append(file)
PSO3_df = pd.concat(get_all_data(PSO3_df_list))
my_df_list = []
for i in my_dict['PSO3']:
    temp = PSO3_df[(PSO3_df['GeneName1'].str.match(i)) | (PSO3_df['GeneName2'].str.match(i))]
    for my_index in temp.index:
        if temp.at[my_index,'GeneName1'] != i:
            
            temp.at[my_index,'GeneName2'] = temp.at[my_index,'GeneName1']
            temp.at[my_index,'GeneName1'] = i
            temp_value = temp.at[my_index,'GeneExpr2']
            temp.at[my_index,'GeneExpr2'] = temp.at[my_index,'GeneExpr1']
            temp.at[my_index,'GeneExpr1'] = temp_value
    my_df_list.append(temp)
df = pd.concat(my_df_list)
df
df.to_csv("PSO3.s.table.csv",index=False)
```

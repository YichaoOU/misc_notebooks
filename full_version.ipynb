```python
# hypergenom test
gene2go = {}
go2gene = {}
with open("Hv_IBSC_PGSB_r1_functionalDescriptions_HighConf_GOs.tab") as f:
    for line in f:
        line = line.strip().split()
        gene = line[0].split(".")[0]
        go_anno = line[-1]
        
        if not gene2go.has_key(gene):
            gene2go[gene] = []
        gene2go[gene].append(go_anno)
        if not go2gene.has_key(go_anno):
            go2gene[go_anno] = []
        go2gene[go_anno].append(gene)        
            
```

```python
from scipy.stats import hypergeom
out = open("GO_enrichment_test_result.csv","wb")
total_genes = len(gene2go)

outline = ['cluster_id','number_of_genes','enriched_GO','p_value','overlapped_number_of_genes','total_genes_in_the_term']
print >>out,",".join(outline)
for cluster_id in my_clusters.assignment.unique().tolist():
    genes = my_clusters.index[my_clusters.assignment == cluster_id].tolist()
    number_of_genes = len(genes)
    print cluster_id,number_of_genes
    for term in go2gene:
        current_go_genes = go2gene[term]
        overlapping_genes = set(current_go_genes).intersection(genes)
        overlapped_number_of_genes = len(overlapping_genes)
        if overlapped_number_of_genes < 2:
            continue
        p_value = hypergeom.sf(overlapped_number_of_genes-1, total_genes, len(current_go_genes), number_of_genes)
        outline = [cluster_id,number_of_genes,term,p_value,overlapped_number_of_genes,len(current_go_genes)]
        print >>out,",".join(map(lambda x:str(x),outline))
            
```

```python
def cmap_discretize(cmap, N):
    """Return a discrete colormap from the continuous colormap cmap.

        cmap: colormap instance, eg. cm.jet. 
        N: number of colors.

    Example
        x = resize(arange(100), (5,100))
        djet = cmap_discretize(cm.jet, 5)
        imshow(x, cmap=djet)
    """

    if type(cmap) == str:
        cmap = get_cmap(cmap)
    colors_i = concatenate((linspace(0, 1., N), (0.,0.,0.,0.)))
    colors_rgba = cmap(colors_i)
    indices = linspace(0, 1., N+1)
    cdict = {}
    for ki,key in enumerate(('red','green','blue')):
        cdict[key] = [ (indices[i], colors_rgba[i-1,ki], colors_rgba[i,ki]) for i in xrange(N+1) ]
    # Return colormap object.
    return matplotlib.colors.LinearSegmentedColormap(cmap.name + "_%d"%N, cdict, 1024)
```

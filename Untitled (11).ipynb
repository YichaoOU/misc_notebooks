```python
chrom = "chr13"
start = 21700000
stop = 24100000
mid = (stop+start)/2.0
```

```python
df[['Score','Distance']].max(axis=1)
```

```python
df['W1_distance_to_locus'] = (df.Window1.str.split('-',expand=True)[1].astype(int) - mid)/30000.0
df['W2_distance_to_locus'] = (df.Window2.str.split('-',expand=True)[1].astype(int) - mid)/30000.0
```

```python
thres = (stop-start)/(2*30000.0)
```

```python
df['max']=abs(df[['W1_distance_to_locus','W2_distance_to_locus']]).max(axis=1)
```

```python
for i in df.index:
    w1 = df.at[i,'W1_distance_to_locus']
    w2 = df.at[i,'W2_distance_to_locus']    
    if abs(w2) > abs(w1):
        df.at[i,'max'] = w2
    else:
        df.at[i,'max'] = w1
```

```python
df
```

```python
window_list = df['Window1'].tolist()+df['Window2'].tolist()
window_list = list(set(window_list))
```

```python
res=30000
```

```python
window_dict = {}
for w in window_list:
    position = int(w.split("-")[-1])-res/2
    if position < start-res/2:
        dis = abs(position-start)
        x = int(int(dis/res) / 200 ) + 1
        window_dict[w] = -x
    elif position > stop+res/2:
        dis = abs(position-start)
        x = int(int(dis/res) / 200 ) + 1
        window_dict[w] = x      
    else:
        window_dict[w] = 0    
```

```python
import networkx as nx
G=nx.from_pandas_edgelist(df, 'Window1', 'Window2', edge_attr=['Score','Distance'])
```

```python

nx.set_node_attributes(G, window_dict, "distancebins")

```

```python
nx.write_gml(G, 'test.gml')

```

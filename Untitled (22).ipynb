```r
my_line = function(x){
    0.9986 * x + 0.127 + rnorm(1,0,sqrt(10.98))
}
```

```r
simulation = function(T,v){
    
    current_v = my_line(v)
    my_vector = c(current_v)
    for (i in 2:T){
        current_v = my_line(current_v)
        my_vector = c(my_vector,current_v)
    }
    return (my_vector)
}
```

```r
a1=simulation(100,10)
a2=simulation(100,10)
a3=simulation(100,10)
b1=simulation(100,20)
b2=simulation(100,20)
b3=simulation(100,20)
```

```r
library(ggplot2)
```

```r
df = data.frame(time = seq(100),a1 = a1,a2 = a2,a3=a3,b1=b1,b2=b2,b3 = b3)
```

```r
ggplot(df, aes(time)) + 
  geom_line(aes(y = a1, colour = "y_0=10")) + 
  geom_line(aes(y = a2, colour = "y_0=10")) + 
  geom_line(aes(y = a3, colour = "y_0=10")) + 
  geom_line(aes(y = b1, colour = "y_0=20")) +
  geom_line(aes(y = b2, colour = "y_0=20")) + 
  geom_line(aes(y = b3, colour = "y_0=20")) +
ylab("price")
```

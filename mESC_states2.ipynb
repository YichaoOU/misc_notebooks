```python
regulators = ["Pou5","Sox","Nanog","Klf","Tcf","Gata","Twist","Dppa","Lhx","Nr2","Col2","Hox"]
```

```python
regulators = list(map(lambda x:x.lower(),regulators))
regulators
```

```python
cols = ["R1-Up-Genes","R2-Up-Genes","R1-Down-Genes","R2-Down-Genes","R1-Overlap-Genes","R2-Overlap-Genes"]
```

```python
import pandas as pd
gam = pd.read_csv("chromatin_interaction_pairs_of_features.csv")
gam.head()
```

```python
names = []
for c in cols:
    for r in regulators:
        try:
            names += gam[c][gam[c].str.match(r,case=False) == True].tolist()
        except:
            continue
```

```python
genes = []
for i in names:
    temp = i.split(",")
#     print (temp)
    for r in regulators:
        for t in temp:
#             print (t)
            t = t.replace(" ","")
            if r in t.lower():
#                 print (t)
                genes.append(t)
        
            
```

```python
genes = list(set(genes))
```

```python
genes
```

```python
test = pd.read_csv("F17F4.GAM05.csv")
```

```python
test.head()
```

```python
index_list = []
for c in cols:
    for g in genes:
        try:
            index_list += test[test[c].str.match(g,case=False) == True].index.tolist()
        except:
            continue
list(set(index_list))
```

```python
res = pd.DataFrame(columns = genes+['total_number_interactions'])
```

```python
res
```

```python
import glob
files = glob.glob("*GAM05.csv")
```

```python
for f in files:
    label = f.split(".")[0]
    df = pd.read_csv(f)
    shapes = df.shape
    my_row = []
    for g in genes:
        count = []
        for c in cols:
            try:
                count += df[df[c].str.match(g,case=False) == True].index.tolist()
            except:
                continue
        my_row.append(len(list(set(count))))
    my_row.append(shapes[0])
    res.loc[label] = my_row
```

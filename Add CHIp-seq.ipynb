```python
dff['Polycomb'] = df3['H3K27me3.mean'].loc[df.index]
```

```python
# dff["Imotif","Hmotif","E4motif","PCGmotif","Pmotif","E8motif"]
dff = pd.concat([dff,df3[["Imotif","Hmotif","E4motif","PCGmotif","Pmotif","E8motif"]].loc[df.index]],axis=1)
```

```python
def gene_exp_feature(x):
    for i in x.split(","):
        try:
            if float(i.split(": ")[-1]) >=1:
                return 1
        except:
            continue
    return 0
dff['Active_gene'] = dff['Gene expr'].apply(lambda x:gene_exp_feature(x))
```

```python
dff.to_csv("table.csv")
```

```python
def poly(x):
    if x>=2:
        return True
    else:
        return False
dff['Polycomb'] = dff['Polycomb'].apply(lambda x:poly(x))
```

```python
def gene1(x):
    if "Hist1" in x:
        return "Hist1"
    if "Vmn" in x:
        return "Vmn"
    if "Btn" in x:
        return "Btn"
    if "Slc17" in x:
        return "Slc17"
    if "Olfr" in x:
        return "Olfr"
    if "Zfp" in x:
        return "Zfp"
    return "Others"
                      
dff['Genes'] = dff['Genes'].apply(lambda x:gene1(x))
```

```python
dff.to_csv("table.csv")
```

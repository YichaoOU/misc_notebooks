```python
def PCA_2d_plot(new_X,Y):
	colors =['red','green']
	plt.figure()
	for color, i, target_name in zip(colors, [0,1], ['neg','pos']):
	    plt.scatter(new_X[Y == i, 0], new_X[Y == i, 1], color=color, alpha=.4, lw=1,
	                label=target_name)
	plt.legend(loc='best', shadow=False, scatterpoints=1)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
my_new_data = pca.fit_transform(df[df.columns[:-1]])
PCA_2d_plot(my_new_data,Y=df[df.columns[-1]])
```

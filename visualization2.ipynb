```python
# calculate DEG
from six import string_types
deg = {}
for k in gset1.keys():
    for g in gset1[k].index.tolist():
        if not isinstance(g, string_types):
            continue
        a1 = gset1[k].at[g,'SL ESC-Day0 (-Dox)']
        a2 = gset1[k].at[g,'EB-Day1 (-Dox)']
        try:
            if len(a1) > 1:
                a1 = a1[0]
                a2 = a2[0]
        except:
            pass
        if a1 > a2:
            deg[g] = (a1-a2)/a2
        else:
            deg[g] = -(a2-a1)/a1
```

```python
for x in sorted(deg,key=lambda x:deg[x]):
    print (x,deg[x])
```

```python
sorted_deg = sorted(deg,key=lambda x:deg[x])
```

```python
UP_in_Naive = sorted_deg[::-1][:100]
UP_in_Poised= sorted_deg[:100]
```

```python
UP_in_Naive = []
UP_in_Poised = []
for x in deg:
    if deg[x] >= 2:
        UP_in_Naive.append(x)
    if deg[x] <= -2:
        UP_in_Poised.append(x)
```

```python
UP_in_Naive_columns = []
for c in df.columns:
    gene = c.split(",")[0]
    if gene in UP_in_Naive:
        UP_in_Naive_columns.append(c)
print (len(UP_in_Naive_columns))
```

```python
UP_in_Poised_columns = []
for c in df.columns:
    gene = c.split(",")[0]
    if gene in UP_in_Poised:
        UP_in_Poised_columns.append(c)
print (len(UP_in_Poised_columns))
```

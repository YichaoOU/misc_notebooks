```python
import pandas as pd
```

```python
df = pd.read_csv("chromHMM_motif_features.csv",index_col=0)
```

```python
df.head()
```

```python
enhancer = ['E4','E8']
promoter = ['E6','E7']
```

```python
PE_type = []
for x in df.index:
    if df.at[x,'state'] in enhancer:
        PE_type.append("enhancer")
    if df.at[x,'state'] in promoter:
        PE_type.append("promoter")
```

```python
pos = df.loc[gene.index]
```

```python
from scipy.stats import ttest_ind
```

```python
for c in pos.columns:
    a = pos[c]
    b = background[c]
    print ttest_ind(a, b, equal_var=False)
```

```python
c = "LHX3_MOUSE.H11MO.0.C"
```

```python
a = pos[c]
b = background[c]
plot_df_pos = pd.DataFrame(columns = ['motif','coverage','class'])
plot_df_pos['motif'] = [c]*len(a)
plot_df_pos['coverage'] = a.tolist()
plot_df_pos['class'] = ["pos"]*len(a)
plot_df_neg = pd.DataFrame(columns = ['motif','coverage','class'])
plot_df_neg['motif'] = [c]*len(b)
plot_df_neg['coverage'] = b.tolist()
plot_df_neg['class'] = ["neg"]*len(b)
plot_df = pd.concat([plot_df_pos,plot_df_neg])
```

```python
count = 0
res = pd.DataFrame(columns=['Motif','ForeCov','BackCov'])
for c in pos.columns:
    a = pos[c]
    b = background[c]
    forecov = (a > 0).sum()/float(len(a))
    backcov = (b > 0).sum()/float(len(b))
    res.loc[count] = [c,forecov,backcov]
    count += 1

    
```

```python
import matplotlib.pyplot as plt
# grid = sns.JointGrid('ForeCov', 'BackCov', data=res, space=0, size=6, ratio=50)
# grid.plot_joint(plt.scatter, color="g")

lm = sns.lmplot('ForeCov', 'BackCov', data=res,fit_reg=False, scatter_kws={'alpha':0.15})
lm.set(ylim=(0, 1.05))
lm.set(xlim=(0, 1.05))
```

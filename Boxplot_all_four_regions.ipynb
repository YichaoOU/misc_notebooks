```python
import pandas as pd
df = pd.read_csv("triplet_motif_ENCODE.csv",index_col = 0)
```

```python
df.shape
```

```python
column_list = df.columns.tolist()
```

```python
df = df/df.max()
```

```python
class_1 = pd.melt(df[df['high'] == 1],value_vars=column_list[4:])
class_1['class'] = "High"
class_2 = pd.melt(df[df['low'] == 1],value_vars=column_list[4:])
class_2['class'] = "Low"
class_3 = pd.melt(df[df['median'] == 1],value_vars=column_list[4:])
class_3['class'] = "Median"
class_4 = pd.melt(df[df['Super_Enh'] == 1],value_vars=column_list[4:])
class_4['class'] = "Super_Enh"
plot_data = pd.concat([class_1,class_2,class_3,class_4])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
f, ax = plt.subplots(figsize=(12, 70))
sns.set(font_scale=1.5)
sns.boxplot(x="value",y="variable",hue="class",data=plot_data,palette=sns.hls_palette(4))
```

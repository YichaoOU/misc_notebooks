```python
enriched_motif_list = []
for x in file_list:
    print (x)
    file = x+"/Hist1_"+x+"_result.enriched_motifs.csv"
    df = pd.read_csv(file)
    print (df[(df.Difference >= 0.1) & (df['P-value'] <= 0.05/357)& (df.State == "E7")])
    motif_list1 = df[(df.Difference >= 0.1) & (df['P-value'] <= 0.05/357) & (df.State == "E7")].Motif.tolist()
    motif_list1 = list(map(lambda x:x.split("_")[0].replace("mm9.",""),motif_list1))  
    result = list(set(motif_list1).intersection(motif_list2))
    enriched_motif_list.append(result)
set(enriched_motif_list[0]).intersection(*enriched_motif_list)
labels = venn.get_labels(enriched_motif_list, fill=['number'])
label_list = list(map(lambda x:x.replace("D_norm","D'"),file_list)) 
fig, ax = venn.venn4(labels, names=label_list)
```

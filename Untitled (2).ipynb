```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Hist1.Dprime.embeddings.csv",header = None, index_col=0)
```

```python
df.describe().to_csv("test.csv")
```

```python
df.head()
```

```python
df = df.set_index(1)
```

```python
from sklearn import manifold
```

```python
my_scaler = manifold.MDS(n_jobs=-1)
MDS_data = my_scaler.fit_transform(df)
```

```python
MDS_data
```

```python
import matplotlib.pyplot as plt
plt.scatter(MDS_data[:, 0], MDS_data[:, 1], alpha=0.2)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df)
X_pca = pca.transform(df)
```

```python
X_pca
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.2)
```

```python
my_scaler = manifold.TSNE()
TSNE_data = my_scaler.fit_transform(df)
plt.scatter(TSNE_data[:, 0], TSNE_data[:, 1], alpha=0.2)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5,n_jobs=-1)
kmeans.fit(TSNE_data)
y_kmeans = kmeans.predict(TSNE_data)
plt.scatter(TSNE_data[:, 0], TSNE_data[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

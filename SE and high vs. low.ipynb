```python
import pandas as pd
df = pd.read_csv("triplet_motif_ENCODE.csv",index_col = 0)
```

```python
df[df['high'] == 1].shape
```

```python
df_se_high = pd.concat([df[df['high'] == 1],df[df['Super_Enh'] == 1]])
```

```python
df_se_high.shape
```

```python
new_df = pd.concat([df[df['high'] == 1],df[df['Super_Enh'] == 1],df[df['low'] == 1]])
```

```python
new_df.shape
```

```python
new_df.head()
```

```python
column_list = df.columns.tolist()
column_list
```

```python
new_df = new_df/new_df.max()
```

```python
class_1 = pd.melt(new_df[new_df['low'] == 0],value_vars=column_list[4:])
```

```python
class_1['class'] = "SE/H"
class_1.head()
```

```python
class_2 = pd.melt(new_df[new_df['low'] == 1],value_vars=column_list[4:])
class_2['class'] = "low"
class_2.head()
```

```python
plot_data = pd.concat([class_1,class_2])
plot_data.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
f, ax = plt.subplots(figsize=(12, 48))
sns.set(font_scale=2)
sns.boxplot(x="value",y="variable",hue="class",data=plot_data,palette=sns.hls_palette(2))
# sns.swarmplot(x="value", y="variable",hue="class", data=plot_data,split =True,palette=sns.hls_palette(2))
```

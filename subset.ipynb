```python
from Bio import SeqIO as io
from Bio.Seq import Seq
from joblib import Parallel, delayed
import gc
def read_fasta(file):
	
	seq_hash = {}
	f=open(file,"rU")
	record = io.parse(f,"fasta")
	for r in record:
		seq_hash[str(r.id)]=str(r.seq).upper()
	return seq_hash 
```

```python
my_seq = read_fasta('sequence.fa')
```

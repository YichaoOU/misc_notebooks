```python
import pandas as pd
```

```python
hist1_gene_list = []
```

```python
with open("mm9_gene_ensembl_chr.bed.sorted") as f:
    for line in f:
        line = line.strip().split()
        if "Hist1" in line[-1]:
            hist1_gene_list.append(line[-1])
```

```python
len(hist1_gene_list)
```

```python
def parse_gene_gene_df(file,hist1_gene_list):
    res = {}
    for i in hist1_gene_list:
        res[i] = {}
        for j in hist1_gene_list:
            res[i][j] = 0
    df = pd.read_csv(file,index_col=0)
    for i in df.index:
        gene1 = df.at[i,'gene1'].split(",")
        gene2 = df.at[i,'gene2'].split(",")
        temp=list(set(gene1+gene2))
        for g1 in temp:
            for g2 in temp:
                if g1 in res:
                    if g2 in res[g1]:
                        res[g1][g2] = 1
    return res        

    
```

```python
res=parse_gene_gene_df("Dnorm_hist1_gene_gene_post/Dnorm_hist1_gene_gene_post.GG_AND.csv",hist1_gene_list)
```

```python
res=pd.DataFrame(res)
```

```python
res=res[hist1_gene_list]
```

```python
res=res.loc[hist1_gene_list]
```

```python
import numpy as np
import gc
data = np.load('Network.npy')
data = 1 - data
np.fill_diagonal(data,0)
import pandas as pd
genes = map(lambda x:x.strip().split()[0],open("Genes.txt").readlines())
len(genes)
df = pd.DataFrame(data,dtype=np.float64)
df.index = genes
df.columns = genes
df.head()
del data
gc.collect()
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
k=3
cluster_km = KMeans(n_clusters=k,max_iter=100,precompute_distances=False,verbose=50,n_jobs=2,random_state=0)
cluster_km.fit(df)
```

```python
predicted=cluster_km.predict(df)
from sklearn.metrics import silhouette_samples, silhouette_score
silhouette_avg = silhouette_score(df, predicted)
```

```python
def get_end(name):
    temp = name.split("-")[-1]
    return int(temp)
```

```python
my_index_list = []
for i in pos.index:
    first_name = pos.at[i,'name']
    end = get_end(first_name)
    if end < 23964851 and end > 21715763:
        my_index_list.append(i)
        continue
    first_name = pos.at[i,'variable']
    end = get_end(first_name)
    if end < 23964851 and end > 21715763:
        my_index_list.append(i)
        continue    
```
